$$Mean Filter$$

```python
#Mean Filter
img = cv2.imread('/content/drive/MyDrive/Einstein.jpg')
img = cv2.resize(img, (256,256)) #Resizing the image

#Defining 3X3 filter
filt = np.ones((3,3))

#Getting shape of Image and Kernel
S = img.shape #Image
F = filt.shape #Kernel

#Converting Image into Binary form
imgbit = img.astype('uint8')
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Displaying Noisy Image in 8-bit form
plt.title('Original 8-bit converted Image')
plt.imshow(img)

#Zero Padding
R = S[0]+F[0]-1 #For Row
C = S[1]+F[1]-1 #For Column
Z = np.zeros((R,C))
mean1 = np.zeros((256,256))

#Performing Operations for calculation of Mean

for i in range(S[0]):#for row
  for j in range(S[1]): #for column
    Z[i+np.int((F[0]-1)/2), j+np.int((F[1]-1)/2)] = imgGray[i,j]

print(Z) #Zero Padded Image

#Performing SOP
for  i in range(S[0]):
  for j in range(S[1]):
    k = Z[i:i+F[0], j:j+F[1]]
    l = np.sum(k*filt)
    mean1[i,j] = l




#Displaying filtered image

plt.subplot(1,2,1)
plt.imshow(imgbit, cmap = 'gray')
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(mean1, cmap='gray')
plt.title("Mean Filtered Image")
```

```python
#Applying Mean Filter of size 5X5
filt = np.ones((5,5))

#Getting shape of Image and Kernel
S = img.shape #Image
F = filt.shape #Kernel

#Zero Padding
R = S[0]+F[0]-1 #For Row
C = S[1]+F[1]-1 #For Column
Z = np.zeros((R,C))
mean2 = np.zeros((256,256))

#Performing Operations for calculation of Mean

for i in range(S[0]):#for row
  for j in range(S[1]): #for column
    Z[i+np.int((F[0]-1)/2), j+np.int((F[1]-1)/2)] = imgGray[i,j]

print(Z) #Zero Padded Image


for  i in range(S[0]):
  for j in range(S[1]):
    k = Z[i:i+F[0], j:j+F[1]]
    l = np.sum(k*filt)
    mean2[i,j] = l

#Displaying filtered image

plt.subplot(1,2,1)
plt.imshow(imgbit, cmap = 'gray')
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(mean2, cmap='gray')
plt.title("Mean Filtered Image")

```

```python
#Applying Mean Filter of size 7X7
filt = np.ones((7,7))

#Getting shape of Image and Kernel
S = img.shape #Image
F = filt.shape #Kernel


#Zero Padding
R = S[0]+F[0]-1 #For Row
C = S[1]+F[1]-1 #For Column
Z = np.zeros_like((R,C))
mean3 = np.zeros((256,256))

#Performing Operations for calculation of Mean

for i in range(S[0]):#for row
  for j in range(S[1]): #for column
    Z[i+np.int((F[0]-1)/2), j+np.int((F[1]-1)/2)] = imgGray[i,j]

print(Z) #Zero Padded Image


for  i in range(S[0]):
  for j in range(S[1]):
    k = Z[i:i+F[0], j:j+F[1]]
    l = np.sum(k*filt)
    mean3[i,j] = l

#Displaying filtered image

plt.subplot(1,2,1)
plt.imshow(imgbit, cmap = 'gray')
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(mean3, cmap='gray')
plt.title("Mean Filtered Image")

```

$$Median Filters$$

```python
#Applying Median filter
img = cv2.imread('/content/drive/MyDrive/Einstein.jpg')
img = cv2.resize(img, (256,256)) #Resizing the image

#For this, we need an empty kernel
#Defining zero matrix of size 3X3
filt = np.ones((3,3))

#Getting shape of Image and Kernel
S = img.shape #Image
F = filt.shape #Kernel


imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #binary conversion
plt.imshow(imgGray)
plt.imshow(img)

#Zero Padding
R = S[0]+F[0]-1 #For Row
C = S[1]+F[1]-1 #For Column
Z = np.zeros((R,C))
med1 = np.zeros((256,256))

#Performing Operations for calculation of Median

for i in range(S[0]):#for row
  for j in range(S[1]): #for column
    Z[i+np.int((F[0]-1)/2), j+np.int((F[1]-1)/2)] = imgGray[i,j]

print(Z) #Zero Padded Image


for  i in range(S[0]):
  for j in range(S[1]):
    k = Z[i:i+F[0], j:j+F[1]]
    l = np.median(k)
    med1[i,j] = l

#Displaying filtered image

plt.subplot(1,2,1)
plt.imshow(imgGray, cmap = 'gray')
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(med1, cmap='gray')
plt.title("Median Filtered Image")

```

```python
#Applying Median filter of size 5X5


#Defining matrix of size 5X5
filt = np.ones((5,5))

#Getting shape of Image and Kernel
S = img.shape #Image
F = filt.shape #Kernel

#Zero Padding
R = S[0]+F[0]-1 #For Row
C = S[1]+F[1]-1 #For Column
Z = np.zeros((R,C))
med2 = np.zeros((256,256))


#Performing Operations for calculation of Median

for i in range(S[0]): #for row
  for j in range(S[1]): #for column
    Z[i+np.int((F[0]-1)/2), j+np.int((F[1]-1)/2)] = imgGray[i,j]

print(Z) #Zero Padded Image

for  i in range(S[0]):
  for j in range(S[1]):
    k = Z[i:i+F[0], j:j+F[1]]
    l = np.median(k)
    med2[i,j] = l


#Displaying filtered image

plt.subplot(1,2,1)
plt.imshow(imgGray, cmap = 'gray')
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(med2, cmap='gray')
plt.title("Median Filtered Image")
```

```python
#Median filter of size 7X7


#Defining matrix filter of size 5X5
filt = np.ones((7,7))

#Getting shape of Image and Kernel
S = img.shape #Image
F = filt.shape #Kernel

#Zero Padding

R = S[0]+F[0]-1 #For Row
C = S[1]+F[1]-1 #For Column
Z = np.ones((R,C))
med3 = np.ones((256,256))


#Performing Operations for calculation of Median

for i in range(S[0]): #for row
  for j in range(S[1]): #for column
    Z[i+np.int((F[0]-1)/2), j+np.int((F[1]-1)/2)] = imgGray[i,j]

print(Z) #Zero Padded Image

for  i in range(S[0]):
  for j in range(S[1]):
    k = Z[i:i+F[0], j:j+F[1]]
    l = np.median(k)
    med3[i,j] = l


#Displaying filtered image

plt.subplot(1,2,1)
plt.imshow(imgGray, cmap = 'gray')
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(med3, cmap='gray')
plt.title("Median Filtered Image")
```
